* MAPE 측정
- MAPE란 무엇인가?
  MAPE(Mean Absulute Percentage Error)는 예측 모델의 정확도를 평가하기 위한 지표이다. MAPE는 예측 값과 실제 값 간의 차이를 백분율로 나타내어, 예측 오류의 크기를 직관적으로 파악할 수 있게해준다
- MAPE는 낮은 값이 예측 값과 실제 값이 가까움을 의미하고 높은 값은 실제 값에서 벗어났음을 의미한다. 즉, 낮을 수록 모델의 일반화 성능이 좋은 것이다.

```python
csv_names = ["하이테크센터", "로스쿨관", "1호관(본관)", "7호관(학생회관)", 
             "2호남관", "5호북관", "2호북관", "김현태인하드림센터", "5호동관",
             "5호남관", "인하드림센터", "9호관", "60주년기념관", "서호관"]
getcontext().prec = 28
```

```python
fedmut_predict = pd.read_csv("predictions(fedmut_previous_day).csv", header=None, dtype=str)
fedmut_predict = fedmut_predict.applymap(Decimal)
fedmut_predict
```

```python
# Function to calculate MAPE
def calculate_mape(actual, predicted):
    return (sum(abs(a - p) / abs(a) for a, p in zip(actual, predicted)) / len(actual)) * 100
```

```python
fedavg_predict = pd.read_csv("predictions(fedavg_previous_day).csv", header=None, dtype=str)
fedavg_predict = fedavg_predict.applymap(Decimal)
fedavg_predict
```

```python
actual_dataset = []
for i in range(len(csv_names)):
    # Load the actual values
    actual_file_path = f'GeneratedCSV/Preprocessed_{csv_names[i]}.csv'
    actual_data = pd.read_csv(actual_file_path, dtype={'PowerUsage': str})

    # Convert DateTime to datetime format
    actual_data['DateTime'] = pd.to_datetime(actual_data['DateTime'])
    actual_data['PowerUsage'] = actual_data['PowerUsage'].apply(Decimal)
    actual_dataset.append(actual_data)
```
