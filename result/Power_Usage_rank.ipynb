```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_names = ["하이테크센터", "로스쿨관", "1호관(본관)", "7호관(학생회관)", 
                "2호남관", "5호북관", "2호북관", "김현태인하드림센터", "5호동관",
                "5호남관", "인하드림센터", "9호관", "60주년기념관", "서호관"]
```

```python

Hightech = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[0]}.csv")
lawschool = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[1]}.csv")
building1 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[2]}.csv")
building7 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[3]}.csv")
south2 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[4]}.csv")
north5 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[5]}.csv")
north2 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[6]}.csv")
kimhyuntae = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[7]}.csv")
east5 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[8]}.csv")
south5 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[9]}csv")
inhadream = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[10]}.csv")
building9 = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[11]}.csv")
sixty = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[12]}.csv")
west = pd.read_csv(f"GeneratedCSV/Preprocessed_{csv_names[13]}.csv")
```

```python
Hightech.reset_index(drop=True, inplace=True)
north2.reset_index(drop=True, inplace=True)
south2.reset_index(drop=True, inplace=True)
building1.reset_index(drop=True, inplace=True)
east5.reset_index(drop=True, inplace=True)
north5.reset_index(drop=True, inplace=True)
south5.reset_index(drop=True, inplace=True)
building7.reset_index(drop=True, inplace=True)
building9.reset_index(drop=True, inplace=True)
sixty.reset_index(drop=True, inplace=True)
kimhyuntae.reset_index(drop=True, inplace=True)
inhadream.reset_index(drop=True, inplace=True)
lawschool.reset_index(drop=True, inplace=True)
west.reset_index(drop=True, inplace=True)
```

```python
rank = []
```

```python
rank.append((Hightech.loc[len(Hightech)-1, "DataValue"], "Hightech"))
rank.append((north2.loc[len(north2)-1, "DataValue"], "north2"))
rank.append((south2.loc[len(south2)-1, "DataValue"],"south2"))
rank.append((building1.loc[len(building1)-1, "DataValue"], "building1"))
rank.append((east5.loc[len(east5)-1, "DataValue"],"east5"))
rank.append((north5.loc[len(north5)-1, "DataValue"], "north5"))
rank.append((south5.loc[len(south5)-1, "DataValue"], "south5"))
rank.append((building7.loc[len(building7)-1, "DataValue"], "building7"))
rank.append((building9.loc[len(building9)-1, "DataValue"], "building9"))
rank.append((sixty.loc[len(sixty)-1, "DataValue"], "sixty"))
rank.append((kimhyuntae.loc[len(kimhyuntae)-1, "DataValue"], "kimhyuntae"))
rank.append((inhadream.loc[len(inhadream)-1, "DataValue"], "inhadream"))
rank.append((lawschool.loc[len(lawschool)-1, "DataValue"], "lawschool"))
rank.append((west.loc[len(west)-1, "DataValue"], "west"))
```

```python
rank
```

```python
# 데이터 분리
values, labels = zip(*rank)


# 막대 그래프 생성
plt.figure(figsize=(14, 7))
bars = plt.bar(labels, values, color='dodgerblue')

# 그래프 디자인
plt.xticks(rotation=0)
plt.xlabel('Buildings')
plt.ylabel('Power Usage')
plt.title('Power Usage by Building')
plt.tight_layout()

# 그래프 표시
plt.show()
```
